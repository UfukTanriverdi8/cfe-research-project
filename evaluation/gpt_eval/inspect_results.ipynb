```python
import pandas as pd
import json
import numpy as np


df = pd.read_csv("../../data/eval_results/gpt/100q_with_gpt_f1_scores.csv")
```

```python
# first sort by gpt_conf_score descending
# for each row show the info of simple_conf, adv_conf and gpt_conf_score > 50
df = df.sort_values(by='gpt_conf_score', ascending=False)
for i, row in df.iterrows():
    print(f"Row {i}:")
    print(f"Question: {row['question']}")
    print(f"Simple Conf: {row['simple_conf']}")
    print(f"Adv Conf: {row['adv_conf']}")
    print(f"GPT Conf:{row['gpt_conf_score'] > 50} | Score: {row['gpt_conf_score']}")
    print(f"Mentioned Candidates: {row['mentioned_cands']}")
    print(f"Ideal Candidates: {row['gpt_cands']}")
    print(f"GPT Candidates Intersection: {row['gpt_cands_intersection']}")
    print("--------------------------------------------------")
```

```python
# correlation between simple_conf and gpt_conf_score > 50
df['gpt_conf_above_50'] = df['gpt_conf_score'] > 50
simple_vs_gpt = df['simple_conf'].corr(df['gpt_conf_above_50'].astype(int))
print(f"Correlation between simple_conf and gpt_conf_score > 50: {simple_vs_gpt}")
# correlation between adv_conf and gpt_conf_score > 50
adv_vs_gpt = df['adv_conf'].corr(df['gpt_conf_above_50'].astype(int))
print(f"Correlation between adv_conf and gpt_conf_score > 50: {adv_vs_gpt}")
```

```python
# correlation between simple_metric and gpt_f1_score for all gpt_conf_score > 50
df_above_50 = df[df['gpt_conf_score'] > 50]
simple_metric_vs_gpt_f1 = df_above_50['simple_metric'].corr(df_above_50['gpt_f1_score'])
print(f"Correlation between simple_metric and gpt_f1_score for gpt_conf_score > 50: {simple_metric_vs_gpt_f1}")
# correlation between adv_metric and gpt_f1_score for all gpt_conf_score > 50
adv_metric_vs_gpt_f1 = df_above_50['adv_metric'].corr(df_above_50['gpt_f1_score'])
print(f"Correlation between adv_metric and gpt_f1_score for gpt_conf_score > 50: {adv_metric_vs_gpt_f1}")
```
